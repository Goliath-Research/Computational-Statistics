## Uniform Distribution

The **continuous uniform distribution** or rectangular distribution is a family of symmetric probability distributions. 

It describes an experiment where the observed values are equally spread across the range of the data set.

```python
from scipy.stats import uniform
```

- `loc`: is the starting point

- `scale`: is the width

- `size`: is the number of values

- Let's generate 10 values with uniform distribution from 0 to 1.

```python
uniform.rvs(size=10, loc = 0, scale=1)
```

Let's generate 10 values with uniform distribution from 0 to 10

```python
uniform.rvs(size=10, loc = 0, scale=10)
```

Let's generate 10 values with uniform distribution from 0 to 100

```python
uniform.rvs(size=10, loc = 0, scale=100)
```

```python
unf = uniform.rvs(size=10, loc = 0, scale=10)
print(unf)
print(type(unf))
```

`unf` is a NumPy array, so you can calculate some central tendency statistics:

```python
print('The min value is  = %.2f' %(unf.min()))
print('The mean value is = %.2f' %(unf.mean()))
print('The max value is  = %.2f' %(unf.max()))
```

And some spread measures:

```python
print('The variance is           = %.2f' %(unf.var()))
print('The standard deviation is = %.2f' %(unf.std()))
print('The range is              = %.2f' %(unf.max()-unf.min()))
```

Generating 10000 values:

```python
data_uniform1 = uniform.rvs(loc=0, scale=10, size=10000)
```

```python
ax = sns.histplot(x=data_uniform1, kde=False, color='darkblue')
ax.set(xlabel='Uniform ', ylabel='Frequency');
```

Changing the `bin` parameter:

```python
ax = sns.histplot(x=data_uniform1, bins=10, kde=False, color='darkblue')
ax.set(xlabel='Uniform ', ylabel='Frequency');
```

Including a density estimation:

```python
ax = sns.histplot(data_uniform1, kde=True, color='darkblue')
ax.set(title='Uniform Distribution');
```

```python
ax = sns.histplot(data_uniform1, bins=10, kde=True, color='darkblue')
ax.set(title='Uniform Distribution');
```

Generating a uniform distribution between 20 and 40:

```python
data_uniform2 = uniform.rvs(loc=20, scale=20, size=10000)
```

```python
ax = sns.histplot(x=data_uniform2, kde=True, color='darkgreen')
ax.set(title='Uniform Distribution');
```

Generating a uniform distribution between 10 and 60

```python
data_uniform3 = uniform.rvs(loc=10, scale=50, size=10000)
```

```python
ax = sns.histplot(x=data_uniform3, kde=True, color='red')
ax.set(title='Uniform Distribution');
```

Plotting all together

```python
plt.figure(figsize=(10,5))
ax = sns.kdeplot(x=data_uniform3, shade=True, color='red', label='U(10,80)')
ax = sns.kdeplot(x=data_uniform2, shade=True, color='darkgreen', label='U(20,40)')
ax = sns.kdeplot(x=data_uniform1, shade=True, color='blue', label='U(0,10)')                  
ax.set(title='Uniform Distributions');
plt.legend();                  
```

```python
plt.figure(figsize=(10,5))
ax = sns.kdeplot(x=data_uniform3, color='red', label='U(10,80)')
ax = sns.kdeplot(x=data_uniform2, color='darkgreen', label='U(20,40)')
ax = sns.kdeplot(x=data_uniform1, color='blue', label='U(0,10)')                  
ax.set(title='Uniform Distributions');
plt.legend();  
```

## Normal Distribution

`Normal distribution`, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean. 

In graph form, normal distribution will appear as a bell curve.

```python
from scipy.stats import norm
```

Generating 10 values from N(0,1)

```python
n1 = norm.rvs(loc=0, scale=1, size=10)
print(n1)
print('\n[Min,Max] = [%.2f, %.2f]' %(n1.min(),n1.max()))
print(type(n1))
```

Generating 10 values from N(0,10)

```python
n2 = norm.rvs(loc=0,scale=10,size=10)
print(n2)
print('\n[Min,Max] = [%.2f,%.2f]' %(n2.min(),n2.max()))
print(type(n2))
```

Generating 10 values from N(0,100)

```python
n3 = norm.rvs(loc=0,scale=100,size=10)
print(n3)
print('\n[Min,Max] = [%.2f,%.2f]' %(n3.min(),n3.max()))
print(type(n3))
```

Generating 10000 values

```python
# N(0,1)
data_normal1 = norm.rvs(loc=0,scale=1,size=10000)
```

```python
ax = sns.histplot(x=data_normal1, bins=100, kde=True, color='blue')
ax.set(title='Normal Distribution');
```

```python
# N(5,2)
data_normal2 = norm.rvs(loc=5,scale=2,size=10000)
```

```python
ax = sns.histplot(x=data_normal2, bins=100, kde=True, color='darkgreen')
ax.set(title='Normal Distribution');
```

```python
# N(10,5)
data_normal3 = norm.rvs(loc=10,scale=5,size=10000)
```

```python
ax = sns.histplot(x=data_normal3, bins=100, kde=True, color='red')
ax.set(title='Normal Distribution');
```

```python
# Plotting all together
plt.figure(figsize=(10,5))
ax = sns.kdeplot(x=data_normal1, shade=True, color='blue', label='N(0,1)')
ax = sns.kdeplot(x=data_normal2, shade=True, color='darkgreen', label='N(5,4)')
ax = sns.kdeplot(x=data_normal3, shade=True, color='red', label='N(10,25)')
ax.set(title='Normal Distributions');
plt.legend(); 
```

```python
plt.figure(figsize=(10,5))
ax = sns.kdeplot(x=data_normal1, color='blue', label='N(0,1)')
ax = sns.kdeplot(x=data_normal2, color='darkgreen', label='N(5,4)')
ax = sns.kdeplot(x=data_normal3, color='red', label='N(10,25)')
ax.set(title='Normal Distributions');
plt.legend(); 
```

### Exercise

An admission test for college X is roughly normally distributed with a mean of 600 and a standard deviation of 100.
- a) What is the probability of a student scores above 550?
- b) What is the probability of a student scores less than 800?
- c) What is the probability of a student scores between 500 and 800?

```python
test = norm(600,100)
```

```python
print('P(Test score >= 500) = %.3f' %(1 - test.cdf(550)))
```

```python
print('P(Test score < 800) = %.3f' %(test.cdf(800)))
```

```python
print('P(500 <= Test score <= 800) = %.3f' 
        %(test.cdf(800) - test.cdf(500)))
```

## Exponential Distribution

**Exponential distribution** is the probability distribution of the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate.

The distribution is supported on the interval `[0, ∞)`

The probability density above is defined in the “standardized” form. To shift and/or scale the distribution use the `loc` and `scale` parameters. Specifically, `expon.pdf(x, loc, scale)` is identically equivalent to `expon.pdf(y) / scale` with `y = (x - loc) / scale`.

```python
from scipy.stats import expon
```

Generating 10 values from `Exp(1)`

```python
e1 = expon.rvs(loc=0,scale=1,size=10)
print(e1)
print('\n[Min,Max] = [%.2f,%.2f]' %(e1.min(),e1.max()))
print(type(e1))
```

```python
print('The mean value is         = %.2f' %(e1.mean()))
print('The variance is           = %.2f' %(e1.var()))
print('The standard deviation is = %.2f' %(e1.std()))
```

Generating 10 values from `Exp(10)`

```python
e2 = expon.rvs(loc=10,scale=1,size=10)
print(e2)
print('\n[Min,Max] = [%.2f,%.2f]' %(e2.min(),e2.max()))
print(type(e2))
```

```python
print('The mean value is         = %.2f' %(e2.mean()))
print('The variance is           = %.2f' %(e2.var()))
print('The standard deviation is = %.2f' %(e2.std()))
```

Generating 10 values from `Exp(10)` with `scale=10`

```python
print('The mean value is         = %.2f' %(e3.mean()))
print('The variance is           = %.2f' %(e3.var()))
print('The standard deviation is = %.2f' %(e3.std()))
```
