# Correlation Analysis

## Objectives

- To understand and demonstrate the concept of correlation between two variables using various data visualizations and statistical measures.
- To explore and compute different types of correlation coefficients, including Pearson's, Spearman's, and Kendall's.
- To generate synthetic datasets and apply correlation analysis to observe the relationships between variables.

## Background

Correlation analysis is a method used to evaluate the extent to which two quantitative variables are associated. The degree of association is measured by a correlation coefficient, ranging from -1 to 1, where values close to 1 or -1 signify a strong relationship, and a value close to 0 indicates a lack of correlation. This notebook employs Python libraries such as NumPy, pandas, matplotlib, seaborn, and scipy to perform and visualize correlation analyses.

## Datasets Used

The datasets used in this analysis are synthetic. They are created within the notebook using random number generation methods from the NumPy library. They are tailored to demonstrate different strengths and directions of correlation (strong positive, strong negative, weak negative, and no correlation).

## Correlation Coefficient Definition

It is a quantity measuring the interdependence of two variables.

Correlation refers to the statistical relationship between two variables, and it is how two variables move with one another. 

```python
import numpy as np 
np.random.seed(100) 
import pandas as pd 

import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("whitegrid") 

import scipy.stats as st 
```

### Strong positive correlation

When the value of one variable increases, the value of the other variable increases in a similar way.

```python
# Generating data
size = 30
np.random.seed(size)
x = np.random.uniform(-3, 3, size=size)
r = np.random.uniform( 3, 5, size=size)

y = 1.5 * x + r 
```

```python
# Creating a DataFrame
df = pd.DataFrame({'x':x, 'y_spc':y})
df.head()
```

```python
# Strong Positive Correlation
sns.scatterplot(x=df.x, y=df.y_spc, s=70, color='darkorange').set_title(
                'Strong Positive Correlation');
```

```python
# Strong Positive Correlation with Trendline
sns.scatterplot(x=df.x, y=df.y_spc, s=70, color='darkorange').set_title(
                'Strong Positive Correlation with Trendline')
z = np.polyfit(df.x, df.y_spc, 1)
p = np.poly1d(z)
plt.plot(df.x, p(df.x), color='grey');
```

```python
# Using seaborn library
sns.lmplot(data=df, x='x', y='y_spc').set(title = 'Strong Positive Correlation');
```

### Strong negative correlation

When the value of one variable increases, the value of the other variable tends to decrease.

```python
df['y_snc'] = - 0.9 * x + r  + 30
df.head()
```

```python
# Strong Negative Correlation
sns.scatterplot(x=df.x, y=df.y_snc, s=70, color='darkgreen').set_title(
                'Strong Negative Correlation');
```

```python
# Strong Negative Correlation with Trendline
sns.scatterplot(x=df.x, y=df.y_snc, s=70, color='darkgreen').set_title(
                'Strong Negative Correlation with Trendline');
z = np.polyfit(df.x, df.y_snc, 1)
p = np.poly1d(z)
plt.plot(df.x, p(df.x), color='grey');
```

```python
# Using seaborn library
sns.lmplot(data=df, x='x', y='y_snc').set(title = 'Strong Negative Correlation with Trendline');
```

### Weak Positive Correlation

A weak positive correlation indicates that both variables tend to go up in response to one another, but their relationship is not very strong.

```python
r2 = np.random.uniform(10, 30, size=size)

df['y_wpc'] = 1.7 * x + r2 
df.head()
```

```python
# Weak Positive Correlation
sns.scatterplot(x=df.x, y=df.y_wpc, s=70, color='coral').set_title(
                'Weak Positive Correlation'); 
```

```python
# Weak Positive Correlation with Trendline
sns.scatterplot(x=df.x, y=df.y_wpc, s=70, color='coral').set_title(
                'Weak Positive Correlation');
z = np.polyfit(df.x, df.y_wpc, 1)
p = np.poly1d(z)
plt.plot(df.x, p(df.x), color='grey');
```

```python
# Using seaborn library
sns.lmplot(data=df, x='x', y='y_wpc').set(title = 'Weak Positive Correlation with Trendline');
```

### Weak Negative Correlation

A weak negative correlation indicates that an increase in one variable is associated with a decrease in the other variable, but the relationship is not strong.

```python
df['y_wnc'] = - 1.7 * x + 1.8 * r2
df.head()
```

```python
# Weak Negative Correlation
sns.scatterplot(x=df.x, y=df.y_wnc, s=70, color='forestgreen').set_title(
                'Weak Negative Correlation');
```

```python
# Weak Negative Correlation with Trendline
sns.scatterplot(x=df.x, y=df.y_wnc, s=70, color='forestgreen').set_title(
                'Weak Negative Correlation');
z = np.polyfit(df.x, df.y_wnc, 1)
p = np.poly1d(z)
plt.plot(df.x, p(df.x), color='grey');
```

```python
# Using seaborn library
sns.lmplot(data=df, x='x', y='y_wnc').set(title = 'Weak Negative Correlation with Trendline');
```

### No correlation

```python
df['y_nc'] = np.random.uniform(10, 20, size=size) * r
df.head()
```

```python
# No correlation
sns.scatterplot(x=df.x, y=df.y_nc, s=70, color='dodgerblue').set_title(
                'No Correlation');
```

```python
r5 = np.random.uniform( 3, 5, size=size)

df['y_nc2'] = df.x * df.x + r
df.head()
```

```python
# No correlation
sns.scatterplot(x=df.x, y=df.y_nc2, s=70, color='dodgerblue').set_title(
                'No Correlation');
```

## Correlation Coefficients and Hypothesis Test

The Correlation Coefficient is a statistical concept that helps establish a relation between predicted and actual values obtained in a statistical experiment. 

The calculated value of the correlation coefficient explains the accuracy between the predicted and actual values.

The Correlation Coefficient value always lies between -1 to +1. 

Hypotesis test for correlation coefficient:
- Ho: correlation coefficient == 0 (No correlation)
- Ha: correlation coefficient != 0 (Significant Correlation)


A p-value lesser than 0.05 indicates a significant correlation coefficient.

### Pearson’s Correlation Coefficient

The Pearson Correlation Coefficient shows the linear relationship between two sets of data.

The function `pearson(x,y)` the Pearson correlation coefficient and its p-value.

```python
plt.figure(figsize=(11,15))

plt.subplot(321)
sns.scatterplot(x=df.x, y=df.y_spc, s=50, color='darkorange').set_title(
                'Strong Positive Correlation');

plt.subplot(322)
sns.scatterplot(x=df.x, y=df.y_snc, s=50, color='darkgreen').set_title(
                'Strong Negative Correlation');

plt.subplot(323)
sns.scatterplot(x=df.x, y=df.y_wpc, s=50, color='coral').set_title(
                'Weak Positive Correlation');

plt.subplot(324)
sns.scatterplot(x=df.x, y=df.y_wnc, s=50, color='forestgreen').set_title(
                'Weak Negative Correlation');

plt.subplot(325)
sns.scatterplot(x=df.x, y=df.y_nc, s=50, color='dodgerblue').set_title(
                'No Correlation');

plt.subplot(326)
sns.scatterplot(x=df.x, y=df.y_nc2, s=50, color='dodgerblue').set_title(
                'No Correlation');
```

```python
plt.figure(figsize=(11, 15))

plt.subplot(321)
title = 'Pearson Corr Coeff = %.3f,   p = %.4f' %(st.pearsonr(df.x, df.y_spc))
sns.scatterplot(x=df.x, y=df.y_spc, s=50, color='darkorange').set_title(title);

plt.subplot(322)
title = 'Pearson Corr Coeff = %.3f,   p = %.4f' %(st.pearsonr(df.x, df.y_snc))
sns.scatterplot(x=df.x, y=df.y_snc, s=50, color='darkgreen').set_title(title);

plt.subplot(323)
title = 'Pearson Corr Coeff = %.3f,   p = %.4f' %(st.pearsonr(df.x, df.y_wpc))
sns.scatterplot(x=df.x, y=df.y_wpc, s=50, color='coral').set_title(title);

plt.subplot(324)
title = 'Pearson Corr Coeff = %.3f,   p = %.4f' %(st.pearsonr(df.x, df.y_wnc))
sns.scatterplot(x=df.x, y=df.y_wnc, s=50, color='forestgreen').set_title(title);

plt.subplot(325)
title = 'Pearson Corr Coeff = %.3f,   p = %.4f' %(st.pearsonr(df.x, df.y_nc))
sns.scatterplot(x=df.x, y=df.y_nc, s=50, color='dodgerblue').set_title(title);

plt.subplot(326)
title = 'Pearson Corr Coeff = %.3f,   p = %.4f' %(st.pearsonr(df.x, df.y_nc2))
sns.scatterplot(x=df.x, y=df.y_nc2, s=50, color='dodgerblue').set_title(title);
```

## Spearman’s Rank Correlation Coefficient

The Spearman's rank correlation coefficient is a technique that can be used to summarise the strength and direction (negative or positive) of a relationship between two variables. The result will always be between -1 and 1. 
 
Spearman's rank-order correlation is the nonparametric version of the Pearson correlation. 

Spearman's correlation coefficient measures the strength and direction of association between two ranked variables.

```python
history_grades = np.array([35, 23, 47, 17, 10, 43, 9, 6, 28])
algebra_grades = np.array([30, 33, 45, 23, 8, 49, 12, 4, 31])
physEdu_grades = np.array([15, 23, 27, 17, 30, 14, 19, 36, 28])

grades = pd.DataFrame({'History':history_grades, 'Algebra':algebra_grades, 
                       'Physical Education':physEdu_grades})
print(grades.shape)
grades.head()
```

```python
plt.figure(figsize=(10,3))

plt.subplot(121)
sns.scatterplot(x=grades.History, y=grades.Algebra, s=70, color='salmon').set_title(
                'History and Algebra Grades');
stat, p =st.spearmanr(history_grades, algebra_grades)
plt.title("Spearman's Rank Corr Coeff = %.3f, p = %.4f\n" % (stat, p))

plt.subplot(122)
sns.scatterplot(x=grades.History, y=grades['Physical Education'], s=70, color='cadetblue').set_title(
                'History and Physical Education Grades');
stat, p =st.spearmanr(history_grades, physEdu_grades)
plt.title("Spearman's Rank Corr Coeff = %.3f, p = %.4f\n" % (stat, p));
```

## Kendall’s Rank Correlation Coefficient

Kendall rank correlation coefficient is a non-parametric alternative to Pearson's correlation coefficient. It measures the ordinal association between two columns of ranked data.

It is also the best alternative to Spearman correlation when our sample size is small and has many tied ranks.

```python
plt.figure(figsize=(10,3))

plt.subplot(121)
sns.scatterplot(x=grades.History, y=grades.Algebra, s=70, color='salmon').set_title(
                'History and Algebra Grades')
stat, p =st.kendalltau(history_grades, algebra_grades)
plt.title("Kendall's Rank Corr Coeff = %.3f, p = %.4f\n" % (stat, p));

plt.subplot(122)
sns.scatterplot(x=grades.History, y=grades['Physical Education'], s=70, color='cadetblue').set_title(
                'History and Physical Education Grades')
stat, p =st.kendalltau(history_grades, physEdu_grades)
plt.title("Kendall's Rank Corr Coeff = %.3f, p = %.4f\n" % (stat, p));
```

## Conclusions

Key Takeaways:
- The correlation coefficient is crucial for determining the degree and direction of a linear relationship between two variables.
- Positive correlation coefficients indicate that as one variable increases, the other variable tends to increase. Conversely, negative correlation coefficients indicate that as one variable increases, the other tends to decrease.
- The Pearson correlation coefficient is suitable for measuring the linear correlation between two normally distributed variables.
- Spearman's and Kendall's rank correlation coefficients are non-parametric tests that can be used when the data do not necessarily meet the normality assumption. Kendall's is more robust for smaller samples or data with many tied ranks.
- Data visualization through scatter plots and regression lines is an effective way to assess the relationship between variables.
